## 2.2 Consumer Theory and Demand

### Introductory model
$Individual \: utility = U\underbrace{(q1,\:\:q2,\:\:}_{\text{Goods}} \overbrace{t, \:\:\: m)}^{\text{Tastes & income}}  $

The steepness of the demand curve comes from the combination of substitution and income effects. <br> <br>

---

#### Demand curve has three qualities:
(1) Price intercept (a) <br>
(2) Quantity intercept $\frac{a}{b}$ <br>
(3) Slope (b)

---

#### All other determinants are demand shifters:
(1) Price of other goods (p2) <br>
(2) Income (m) <br>
(3) Tastes (t)

---

#### Relationship between goods and income: 
(1) **Normal:** Higher income -> Higher Demand <br>
(2) **Inferior:** Higher income -> Lower Demand <br>
(3) **Income Neutral:** Higher income -> Same Demand

---

#### Nonfunctional demand: motivated by qualities other than inherent characteristics
(1) Interdependent - based on other people <br>
$\quad$ (a) Bandwagon effect - purchase because others purchased <br>
$\quad$ (b) Snob effect - purchase to be different than others <br>
$\quad$ (c) Veblen / conspicuous consumption effect - purchase to impress, expensive <br>
(2) Speculative - purchase as an investment <br>
(3) Irrational - purchase on a whim

## 2.3 Technology and Costs (Multiple Products)

$W_{L} = Price\:of\:labor$ <br> $W_{K} = Price\:of\:capital$ <br> $q1 = Quantity\: of\: good\: 1$ <br> $q2 = Quantity\: of\: good\: 2$ <br> $T = Technology$

$ Total\:Cost = TC\: (\underbrace{W_{L}, \: W_{K}},\underbrace{ \: q1, \: q2}, \: T) $

We will see **economies of scale** when the long run average cost falls as output increases. <br>
As long as the average cost is less than the marginal cost, we should keep producing. <br>
**Minimum efficient scale** - Mimimum quantity produced to take advantage of economies of scale.


$ AC=\frac{100\quad-\quad\overbrace{20Q + Q^2}^{\text{Minimize this}} \qquad + \overbrace{40\alpha}^{\text{Parameter (Larger = flatter parabola)}}}{4\alpha} $
<br>
<br>

<br>
<br>

$ At\:the\: quantity\: where\: AC \:is\: minimized,\: AC = MC. $

There is **economies of scope** if joint production by a single firm is cheaper than production by two seperate firms. <br>
Why? <br>
(1) Complements in production (using by-products of each other) <br>
(2) Share common inputs (railroads transporting passengers and freight)

## 2.4 Theory of the Firm: "The goal of the firm is to maximize profits."

$ Economic \: profits  \: =  \: total \: revenue\: - \: economic \: costs $

$ Value \: of \: an \: asset \: = \: PV \: of \: the \:stream\: of \:its\: expected\: future\: returns $




 
### Issue: Firms look to maximize in the long run, not just right now
 
##### Solution: Discounting

$\beta = Discount\: factor\: (PV \: of\: 1 \:dollar\: recieved\: next\: period,\: scale\: from\: 0 \:to\: 1) $ <br>
$\quad \beta = .9 \: represents \:  90 \: cents \: today \: equaling \: 1  \: dollar \: tomorrow$ <br>


$ Value(\pi,\infty) = \beta^0 \pi + \beta^1 \pi + ... + \beta^n \pi$

$ Value(\pi,\infty) = \frac{\pi}{1-\beta} $

---

##### Discounting Example


So, the PV of receiving 1 dollar every day when we value 90c today as 1 USD tomorrow is 10 dollars <br>
D = .9 <br>
$\pi$ = 1 dollar <br>
$ Value = \frac{\pi}{1-\beta} = \frac{1}{1-.9} = 10 $

This helps us see how much profit we will be willing to give up today in exchange for additional profit in the future. A higher $\beta$ implies that a greater value is placed on the future.

### 2.4 Boundaries of the firm

**Why set up a firm?** <br>
(1) Horizontal growth - provides greater economies of scale (merger in same industry)<br>
(2) Conglomerate growth - provides greater economies of scope (merger in different industries)<br>
(3) Reduces transaction costs (costs associated with trading)<br>
(4) Vertical growth - buying up the supply chain

**Forces that limit size of firm:**<br>
(1) Merger with a supplier reduces the suppliers flexibility and control -> market inefficiencies<br>
(2) The supplier may be too large to be profitably owned by a single wholesaler<br>
(3) Managerial capacity - costs grow as firm grows

**Costs can be broken down into 2 groups:**<br>
(1) Those that decline as the org grows (eg transaction costs)<br>
(2) Those that grow as the org grows (eg managerial costs)<br>

$ TC = C_{MGT} + C_{MKT} $
